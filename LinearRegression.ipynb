```python
from sklearn.datasets import load_boston
from pandas import DataFrame
import numpy as np
```

```python
boston = load_boston()
```

```python
# 説明変数をデータフレームへ格納
df = DataFrame(data = boston.data, columns = boston.feature_names)
```

```python
# 目的変数をデータフレームへ追加
df['PRICE'] = np.array(boston.target)
```

```python
df.head()
```

```python
df.shape
```

```python
df[['RM']].head()
```

```python
data = df.loc[:, ['RM']].values
```

```python
data[0:10]
```

```python
target = df.loc[:, 'PRICE'].values
```

```python
target[0:10]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(data, target)
```

```python
model.predict([[5]])
```

重回帰分析

```python
df[['CRIM', 'RM']].head()
```

```python
data2 = df.loc[:, ['CRIM', 'RM']].values
```

```python
target2 = df.loc[:, ['PRICE']].values
```

```python
model2 = LinearRegression()
```

```python
model2.fit(data2, target2)
```

```python
model2.predict([[0.3, 4]])
```

```python
print(model2.coef_)
```

```python
#重回帰分析の回帰係数と切片を出力
print('推定回帰係数1：　%.3f, 推定回帰係数２：　%.3f, 推定切片：　%.3f' % (model2.coef_[0,0], model2.coef_[0,1], model2.intercept_))
```

```python
from sklearn.model_selection import train_test_split
```

```python
# データの分割
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.3, random_state = 666)
# 学習用データでパラメータ推定
model.fit(X_train, y_train)
#　作成したモデルから予測（学習用、推論用）
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#　学習用残差プロット
plt.scatter(y_train_pred, y_train_pred - y_train, c = 'blue', marker = 'o', label = 'Train Data')
# 検証用残差プロット
plt.scatter(y_test_pred, y_test_pred - y_test, c = 'lightgreen', marker = 's', label = 'Test Data')
#ラベルプロット
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
#凡例プロット
plt.legend(loc = 'upper left')
# Y=０に直線を引く
plt.hlines(y = 0, xmin = -10, xmax = 50, lw = 5, color = 'red')
plt.xlim([10, 50])
plt.show()
```

```python
# MSEを求める
from sklearn.metrics import mean_squared_error
# 学習用検証用データに関して平均２乗誤差を出力
print('MSE Train: %.3f, Test: %.3f' % (mean_squared_error(y_train, y_train_pred), mean_squared_error(y_test, y_test_pred)))
# 決定係数を出力
print('R^2 Train: %.3f, Test: %.3f' % (model.score(X_train, y_train), model.score(X_test, y_test)))
```
